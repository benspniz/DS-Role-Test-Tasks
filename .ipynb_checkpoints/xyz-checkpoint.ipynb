## Import Relevant Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Data

### a. Import the Data Set

```python
# Loading, checking for missing values & if not then displaying
df = pd.read_csv("Iris.csv", index_col ='Id')
df
```

```python
# Checking for missing values
df.isna().sum()
```

There are no missing values.

## Data Visualization

### Cluster Plots

```python
# Splitting the data set according to species
df_setosa = df[df['Species']=='Iris-setosa']
df_versicolor = df[df['Species']=='Iris-versicolor']
df_virginica = df[df['Species']=='Iris-virginica']
```

```python
# Method to generate the scatter plot
def scatterPlot(df, x_parameter, y_parameter, color_data, marker_data, label_data):
    plt.xlabel(x_parameter)
    plt.ylabel(y_parameter)
    plt.scatter(df[x_parameter], df[y_parameter], color=color_data, marker=marker_data, label=label_data)
    plt.legend(loc='best')
```

```python
# Plotting to view clusters made by different species on the basis of Sepal properties
scatterPlot(df_setosa, 'SepalLengthCm', 'SepalWidthCm', 'red', '+', 'setosa')
scatterPlot(df_versicolor, 'SepalLengthCm', 'SepalWidthCm', 'blue', '.', 'versicolor')
scatterPlot(df_virginica, 'SepalLengthCm', 'SepalWidthCm', 'green', '*', 'virginica')
```

Here we can see a clear distinction between setosa in comparison to the other two species

```python
# Plotting to view clusters made by different species on the basis of Petal properties
scatterPlot(df_setosa, 'PetalLengthCm', 'PetalWidthCm', 'red', '+', 'setosa')
scatterPlot(df_versicolor, 'PetalLengthCm', 'PetalWidthCm', 'blue', '.', 'versicolor')
scatterPlot(df_virginica, 'PetalLengthCm', 'PetalWidthCm', 'green', '*', 'virginica')
```

Here we can see that there is a clear distinction between all three species

### Distribution Plots

```python
# Method to view the distribution of the features
def distributionDataSet(feature):
    plt.subplots(1, 1, sharey=True, figsize =(8,6))
    sns.distplot(df[feature]) 
    plt.xlabel(feature)
```

```python
# Calling the method to view distribution of the features
distributionDataSet('SepalLengthCm')
distributionDataSet('SepalWidthCm')
distributionDataSet('PetalLengthCm')
distributionDataSet('PetalWidthCm')
```

The distributions look "normal". Hence we can use a Gaussian Naive Bayes approach.

## Input and Target

```python
# Creating a Target column from species column using map function
df['Target'] =df.Species.map({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3})
```

```python
# Storing input and target into respective variables
inputs = df.iloc[:,1:-2]
target = df.iloc[:,-1]
```

```python
# Normalizing the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
inputs_scaled = scaler.fit_transform(inputs)
```

### b. Store the split data into separate variables

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_scaled,target,test_size=0.2, random_state = 72)
```

## Model

### c. Create a naive Bayes model on the training dataset

```python
# Importing Gaussian Navive Bayes class
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
# Fitting the model with training data
model.fit(X_train, y_train)
# Training score
print("Train Accuracy: "+str(model.score(X_train, y_train)))
```

### d. Now predict the data model for a testing dataset that you have separated in bit b.

```python
# Displaying the test accuracy
print("Test Accuracy: "+str(model.score(X_test,y_test)))
```

**We were able to achieve a test accuracy of 100% which shows that the model is doing it's job very well!**

My first instinct was to use Grid search so that I can use the entire data set to train the model. But since it was explicitly mentioned under the task to split the data into train and test I followed the simple method depicted above. The code using Grid search is mentioned below.

```python
# Importing Grid Search Cross validation class
from sklearn.model_selection import GridSearchCV
# Instantiating an object for Grid Search Cross validation class and passing Gaussian Navive Bayes model to find the best model
model = GridSearchCV(GaussianNB(),{}, cv=5, return_train_score=False)
# Fitting the model with the entire dataset
model.fit(inputs_scaled, target)
# Storing the results as a dataframe
df_results = pd.DataFrame(model.cv_results_)
#Viewing The best accuracy
print("Accuracy: "+str(model.best_score_))
```
